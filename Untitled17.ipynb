```python
import csv


def read_csv(file_path):
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)
        return [row for row in reader]


def write_csv(file_path, data, fieldnames):
    with open(file_path, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)


data = read_csv('student_performance.csv')


for row in data:
    row["Score"] = str(int(row["Score"]) + 5)


write_csv('updated_student_performance.csv', data, fieldnames=["Student Name", "Student ID", "Subject", "Score", "Attendance Percentage", "Assignment Completion"])

print("Updated data written to 'updated_student_performance.csv'.")
```

```python
import pandas as pd
import re
import os
from functools import wraps


def log_execution(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Executing function: {func.__name__} with arguments {args} and {kwargs}")
        result = func(*args, **kwargs)
        print(f"Execution of function: {func.__name__} completed")
        return result
    return wrapper


def read_data(filename):
    try:

        if not os.path.exists(filename):
            raise FileNotFoundError(f"The file {filename} does not exist.")
        df = pd.read_csv(filename)
        return df
    except FileNotFoundError as e:
        print(e)
        return None
    except pd.errors.ParserError:
        print(f"Error reading the CSV file. It might be corrupted.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None


def validate_student_id(student_id):
    pattern = r'^STU-\d{5}$'
    if re.match(pattern, student_id):
        return True
    else:
        return False


@log_execution
def calculate_aggregate_score(student_name, df):
    try:
        student_data = df[df['Student Name'] == student_name]
        if student_data.empty:
            raise ValueError(f"No data found for student: {student_name}")
        return student_data['Score'].sum()
    except Exception as e:
        print(f"Error in calculate_aggregate_score: {e}")
        return 0


def determine_grade(score):
    try:
        if score >= 90:
            return 'A'
        elif score >= 80:
            return 'B'
        elif score >= 70:
            return 'C'
        else:
            return 'Fail'
    except Exception as e:
        print(f"Error in determine_grade: {e}")
        return 'Error'


@log_execution
def generate_report(student_name, df):
    try:
        total_score = calculate_aggregate_score(student_name, df)
        grade = determine_grade(total_score)
        report = f"Student: {student_name}\nTotal Score: {total_score}\nGrade: {grade}\n"
        return report
    except Exception as e:
        print(f"Error in generate_report: {e}")
        return "Error generating report."

filename = 'student_performance.csv'
df = read_data(filename)

if df is not None:
    students = ['John Doe', 'Jane Smith', 'Alice Johnson', 'Bob Brown']

    for student in students:
        print(generate_report(student, df))
        print("="*50)


    student_ids = ['STU-12345', 'STU-9876', 'STU-67890', '12345']
    for student_id in student_ids:
        if validate_student_id(student_id):
            print(f"{student_id} is a valid student ID.")
        else:
            print(f"{student_id} is an invalid student ID.")
```

```python
import pandas as pd

def add_new_student(data, student_name, student_id, subject, score, attendance, assignment_completion):
    new_student = {
        "Student Name": student_name,
        "Student ID": student_id,
        "Subject": subject,
        "Score": score,
        "Attendance Percentage": attendance,
        "Assignment Completion": assignment_completion
    }
    new_student_df = pd.DataFrame([new_student])
    data = pd.concat([data, new_student_df], ignore_index=True)
    return data

def search_student(data, search_term):
    search_result = data[(data['Student Name'].str.contains(search_term, case=False)) |
                         (data['Student ID'].str.contains(search_term, case=False))]
    return search_result

def update_student(data, student_id, score=None, attendance=None):
    student = data[data['Student ID'] == student_id]
    if student.empty:
        print(f"Student with ID {student_id} not found.")
        return data
    if score is not None:
        data.loc[data['Student ID'] == student_id, 'Score'] = score
    if attendance is not None:
        data.loc[data['Student ID'] == student_id, 'Attendance Percentage'] = attendance
    return data

def interactive_console():
    try:
        data = pd.read_csv("data.csv")
    except FileNotFoundError:
        print("File 'data.csv' not found. Creating a new file.")
        data = pd.DataFrame(columns=["Student Name", "Student ID", "Subject", "Score", "Attendance Percentage", "Assignment Completion"])

    while True:
        print("\nOptions: ")
        print("1. Add new student")
        print("2. Search for a student")
        print("3. Update student's score or attendance")
        print("4. Exit")
        choice = input("Choose an option (1/2/3/4): ")

        if choice == "1":
            student_name = input("Enter student's name: ")
            student_id = input("Enter student ID (e.g., STU-12345): ")
            subject = input("Enter subject: ")
            try:
                score = int(input("Enter score: "))
            except ValueError:
                print("Invalid input for score. Please enter an integer.")
                continue
            try:
                attendance = int(input("Enter attendance percentage: "))
            except ValueError:
                print("Invalid input for attendance. Please enter an integer.")
                continue
            assignment_completion = input("Is the assignment completed? (Yes/No): ")
            data = add_new_student(data, student_name, student_id, subject, score, attendance, assignment_completion)

        elif choice == "2":
            search_term = input("Enter student name or ID to search: ")
            result = search_student(data, search_term)
            if result.empty:
                print("No student found.")
            else:
                print("\nSearch Results:")
                print(result)

        elif choice == "3":
            student_id = input("Enter the student ID to update: ")

            score_input = input("Enter new score (or press Enter to skip): ")
            score = int(score_input) if score_input else None

            attendance_input = input("Enter new attendance percentage (or press Enter to skip): ")
            attendance = int(attendance_input) if attendance_input else None

            data = update_student(data, student_id, score, attendance)

        elif choice == "4":
            data.to_csv("data.csv", index=False)
            print("Data saved to 'data.csv'.")
            break

        else:
            print("Invalid choice. Please try again.")

interactive_console()
```
